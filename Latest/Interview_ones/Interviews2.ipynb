# Precious stone problem

```python
def MaxNoStones(arr, curret_ind, lookup_dict):
  n_rows = len(arr)
  n_col = len(arr[0])

  if curret_ind[0]==0 and curret_ind[1] == n_col-1:
    return arr[0][n_col-1]
  elif curret_ind in lookup_dict:
    return lookup_dict[curret_ind]
  right_num = MaxNoStones(arr, (curret_ind[0], curret_ind[1]+1), lookup_dict) if curret_ind[1]+1 <n_col else 0
  up_num = MaxNoStones(arr, (curret_ind[0]-1, curret_ind[1]), lookup_dict) if curret_ind[0]-1 >=0 else 0

  lookup_dict[curret_ind] =arr[curret_ind[0]][curret_ind[1]] +  max(right_num, up_num)

  return lookup_dict[curret_ind]
```

```python
array = [[2,5,3,5],
       [4,2,7,8],
       [12,0,0,1]]
```

```python
lookup = {}

MaxNoStones(array, (len(array)-1, 0), lookup)
```

```python
def UniqueFunc(a, b):
  """
  a and b are sorted list of integers
  """

  set_a = set(a)
  set_b = set(b)

  unique_a = set_a.difference(set_b)
  unique_b = set_b.difference(set_a)

  common = set_a.intersection(set_b)

  return (sorted(unique_a), sorted(unique_b), sorted(common))
```
