## Top down approach without lookup table

```python
coins = [2, 5, 10, 100]
Amount = 128
```

```python
import sys
def FindCoinChange(coins, Am, counter=[0]):
  """
  coin: array of coins
  Am: amount to be paid using coins
  """
  # coin_list = []
  if Am == 0:
    return 0
  if Am < 0:
    return sys.maxsize  #just a big number to show that combination is possible
  coin_change_array = []
  for coin in coins:
    coin_change_array.append(1 + FindCoinChange(coins, Am-coin, counter))
  # print(coin_change_array)
  counter[0] += 1
  return min(coin_change_array)
```

## Top down approach with lookup table

```python
import sys
def FindCoinChange(coins, Am, lookup_dic, counter=[0]):
  """
  coin: array of coins
  Am: amount to be paid using coins
  """
  # coin_list = []
  if Am == 0:
    return 0
  elif Am in lookup_dic:
    return lookup_dic[Am] 
  if Am < 0:
    return sys.maxsize  #just a big number to show that combination is possible
  coin_change_array = []
  for coin in coins:
    coin_change_array.append(1 + FindCoinChange(coins, Am-coin, lookup_dic, counter))
  # print(coin_change_array)
  counter[0] += 1
  result = min(coin_change_array)
  lookup_dic[Am] = result
  return result
```

```python
coin_list = [1,2,5]
am = 11
num_oper = [0]
lookup = {}
FindCoinChange(coin_list, am, lookup, num_oper)
```

## Bottom down approach

```python
def FindMinCoinChange(coins, amount):
  big_num = 2*amount #big number could be anything bigger than amount
  # ith element of the below array will store the min num of coins needed to make amount = i
  amount_array = [big_num]*(amount+1)
  amount_array[0] = 0 #Since we need 0 coins to form amount=0
  #Now if the final coin used is c, then numcoins(amount) = 1+ numcoins(amount-c)
  for i in range(1, len(amount_array)):
    for coin in coins:
      if i-coin<0:
        continue
      else:
        amount_array[i] = min(1+amount_array[i-coin], amount_array[i])
  return amount_array
```

```python
coin_list = [1,2,5]
am = 11
```

```python
FindMinCoinChange(coin_list, am)
```
