```python
def oper(a,b, opr_str):
  if opr_str =="+":
    return a+b
  if opr_str =="*":
    return a*b
  if opr_str =="-":
    return a-b
```

```python
def evaluate(expression):
  result_vect = []
  oper_strs = set({"*", "+", "-"})
  for ind,el in enumerate(expression):
    if el in oper_strs:
      left_ev = evaluate(expression[:ind])
      right_ev = evaluate(expression[ind+1:])
      result_vect += [oper(int(l),int(r), el) for l in left_ev for r in right_ev]
  if not result_vect:
      return [expression]
  return result_vect
```

```python
evaluate("2-1-1")
```

```python
evaluate("2*3-4*5")
```
