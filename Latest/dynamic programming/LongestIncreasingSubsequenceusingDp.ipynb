```python
def LISLength(arr, i, n, prev, lookup_dic):
  """
  arr- array of numbers
  i - current index we are looking at
  n - number of elements in arr
  prev- previous element in the LIS so far in the recursion
  lookup_dict - {i: LISLentgh(arr, i, n, lookup_dic)}
  """
  # print(prev)
  if i==n:
    return 0
  # elif i in lookup_dic:
  #   return lookup_dic[i]
  #LIS length after excluding the current element
  excl =  LISLength(arr, i+1, n, prev, lookup_dic)

  incl = 0
  #including the current element if it is bigger than teh prev element
  if arr[i] > prev:
    # print(arr[i], prev)
    incl = 1+ LISLength(arr, i+1, n , arr[i], lookup_dic)

  result = max(excl, incl)
  lookup_dic[i] = result
  return result
```
