```python
#Here we are savingthe LCS for each m and n so that we don't have to redo the computations again.
# This gives us  a time complexity of O(n*m) and space complexity of O(n*m) in instead of the
# time complexity of O(n*2^m) in the case of not sorting the values for each m and n. 

def LCS(X, Y, m, n, lookup_dict):
  """
  X, Y- sequences from which we are trying to find the LCS.
  m,n - the indices of X,Y we are currently at.
  lookup_dict: (m,n) - > LCS(m,n)
  """

  if m == 0 or n == 0:
    return 0
  elif (m,n) in lookup_dict:
    return lookup_dict[(m, n)]
  elif X[m-1] == Y[n-1]:
    result = 1+ LCS(X, Y, m-1, n-1, lookup_dict)
    lookup_dict[(m,n)] = result
    return result
  else:
    result = max(LCS(X, Y, m, n-1, lookup_dict), LCS(X, Y, m-1, n, lookup_dict))
    lookup_dict[(m,n)] = result
    return result
```

```python
X = 'ABCBDAB'
Y = 'BDCABA'

# create a dictionary to store solutions to subproblems
lookup = {}

print('The length of the LCS is', LCS(X, Y, len(X), len(Y), lookup))
```
