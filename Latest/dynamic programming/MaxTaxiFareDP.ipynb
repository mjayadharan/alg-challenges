```python
import sys
def FindMaxEarning(rides, n, cur_point, lookup_dic):
  """
  rides list of list of form [[start, end, tip]] as mentioned in description
  n : number of points on the road
  cur_time- current point (0 in the beginning)
  """
  # print("current point: ",cur_point)
  if  cur_point>=n:
    return 0
  elif cur_point in lookup_dic:
    return lookup_dic[cur_point]
  earning_list = [0]
  for ride in rides:
    if ride[0] >= cur_point:
      # print("current rider: ",ride, cur_point)
      ride_earning = (ride[1]-ride[0]+ride[2])+FindMaxEarning(rides, n, ride[1], lookup_dic)
      earning_list.append(ride_earning)
  lookup_dic[cur_point] = max(earning_list)
  return lookup_dic[cur_point]
```

```python
n = 20
rides = [[1,6,1],[3,10,2],[10,12,3],[11,12,2],[12,15,2],[13,18,1]]
lookup = {}
FindMaxEarning(rides, n, 0, lookup)
```
