```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
```

```python
class LinkedList:
  def __init__(self):
    self.head = None
  
  def Insert(self, headref, data):
    new_node = Node(data)
    if headref==None:
      return new_node
    else:
      new_node.next = headref
    headref = new_node
    return headref

  def InsertList(self, headref, data_list):
    for data in data_list:
      headref = self.Insert(headref, data)
    return headref

  def Print(self):
    temp_node = self.head
    ind = 0
    while temp_node:
      print(ind, ":", temp_node.data)
      ind +=1
      temp_node = temp_node.next
```

```python
L = LinkedList()
```

```python
L.head = L.InsertList(L.head, [1,2,3,4,5,6])
```

```python
L.Print()
```

## Finding the middle element of a singly linked list in one go

- Idea is to send two pointers at once, one travelling one node at a time and another one travelling two nodes at a time. By the time the faster pointer reaches the end of the linked list, the slower pointer woudl be at the middle of the linked list. 

```python
def FindMiddleElement(linked_list):
  slower_pointer = linked_list.head
  faster_pointer = linked_list.head
  if faster_pointer:
      faster_pointer = faster_pointer.next

  while(slower_pointer and faster_pointer):
    # print(slower_pointer.data, faster_pointer.data)
    slower_pointer = slower_pointer.next
    faster_pointer = faster_pointer.next
    if faster_pointer:
      faster_pointer = faster_pointer.next
    
  return slower_pointer
```

```python
#Assuming the linked lists are of equal length
from collections import deque
# sum_stack = deque()
def SumLinkedList(headref1, headref2, sum_stack):
  head_1 = headref1
  head_2 = headref2

  if head_1.next == None or head_2.next==None:
    data_sum = head_1.data + head_2.data
    sum_stack.appendleft(data_sum%10)
    # print("reached here")
    return data_sum//10
  
  data_sum = head_1.data + head_2.data + SumLinkedList(head_1.next, head_2.next, sum_stack)
  sum_stack.appendleft(data_sum%10)
  return data_sum//10
```

## Mergin two linked lists that are sorted so the result is also sorte

```python
def mergeLinkedList(l_1, l_2):
  if l_1 == None:
    print("reached here 1")
    # l_1.print()
    return l_2
  elif l_2 == None:
    return l_1
  result = None
  if l_1.data < l_2.data:
    # print("reached here")
    result = l_1
    result.next = mergeLinkedList(l_1.next, l_2)
  else:
    print("reached here 3")
    result = l_2
    result.next = mergeLinkedList(l_1, l_2.next)
  return result
```

```python
l_1 = LinkedList()
l_1.head = l_1.InsertList(l_1.head, [9,7,5,3,1])

l_2 = LinkedList()
l_2.head = l_2.InsertList(l_2.head, [8,6,4,2,0])
```

```python
merged_list = mergeLinkedList(l_1.head, l_2.head)
```

```python
merged_list_2 = LinkedList()
merged_list_2.head = merged_list
```
