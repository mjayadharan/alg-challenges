```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([85.497, 85.481, 85.018, 85.203]).reshape((4,1))
Y = np.array([56.792, 56.027, 54.741, 53.780]).reshape((4,1))
Z = np.array([-19.458, -18.238, -18.282, -17.198]).reshape((4,1))
```

```python
A = np.concatenate((X,Y,Z),axis=1)
A
```

```python
B = []
```

```python
for i in range(3):
    B.append(A[i+1]-A[i])
```

```python
C=np.array(B)
C
```

```python
np.linalg.det(C)
```

```python
#Findin the normals 
norm_1 = np.cross(C[0],C[1])
norm_2 = np.cross(C[1],C[2])
```

```python
print(norm_1)
print(norm_2)
np.dot(norm_1,norm_2)
```
