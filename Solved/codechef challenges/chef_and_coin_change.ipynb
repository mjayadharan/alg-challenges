```python
# cook your dish here
def YesServe(customer_coins):
    coin_counter = {5:0,10:0,15:0}
    for coin in customer_coins:
        need_change= True
        coin_counter[coin]+=1
        change_back = coin - 5
        num_15_in_change = change_back//15
        for num_15 in range(num_15_in_change,-1,-1):
            if not need_change:
                break
            change_back_after_15 = change_back - 15*num_15
            num_10_in_change = change_back_after_15//10
            for num_10 in range(num_10_in_change,-1,-1):
                change_back_after_15_and_10 = change_back_after_15 - 10*num_10
                num_5 = change_back_after_15_and_10//5
                if num_15<=coin_counter[15] and num_10<=coin_counter[10] and num_5<=coin_counter[5]:
                    coin_counter[15]-=num_15 
                    coin_counter[10]-=num_10 
                    coin_counter[5]-=num_5 
                    need_change= False
                    break
        if need_change:
            return "NO"
    
    return "YES"
        

# if __name__=="__main__":
#     T = int(input().strip())
#     for _ in range(T):
#         N = int(input().strip())
#         customer_coins = list(map(int,input().strip().split()))
#         print(YesServe(customer_coins))
        
```
