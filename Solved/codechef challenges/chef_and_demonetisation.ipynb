```python

class CoinChanger:
    def __init__(self,S,N):
        self.N = N if (N%2==0 or N==1) else N-1
        self.S = S
    def min_number_coins(self):
        if S==1:
            return 1
        biggest_coins_num = self.S//self.N
        balance_after_biggest = self.S%self.N
        if balance_after_biggest == 0:
            return biggest_coins_num
        smaller_coins_num = 1 if (balance_after_biggest%2==0 or balance_after_biggest==1) else 2
        return biggest_coins_num+smaller_coins_num

if __name__ == "__main__":
    T = int(input().strip())
    for _ in range(T):
        S,N = [int(el) for el in input().strip().split()]
        coin_changer = CoinChanger(S,N)
        print(coin_changer.min_number_coins())
        
```
