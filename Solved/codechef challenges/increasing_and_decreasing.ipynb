```python
from collections import deque
def check_function(arr):
    n = len(arr)
    if n == 1:
        return (True, arr)
    arr.sort()
    left_deque = deque()
    right_deque = deque()
    for el in arr:
        try:
            if left_deque[-1] != el:
                left_deque.append(el)
            elif right_deque[0] != el:
                right_deque.appendleft(el)
            else:
                return (False, arr)
        except IndexError:
            if not left_deque:
                left_deque.append(el)
            elif not right_deque:
                right_deque.appendleft(el)
    if right_deque and right_deque[0] == left_deque[-1]:
        return (False,arr)
    return (True, left_deque+right_deque)
    

if __name__ == "__main__":
    T = int(input().strip())
    for _ in range(T):
        N = int(input().strip())
        arr = list(map(int,input().strip().split()))
        ans = check_function(arr)
        if ans[0]:
            print("YES")
            print(" ".join(list(map(str,ans[1]))))
        else:
            print("NO")
    
```
