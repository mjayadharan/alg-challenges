Basic idea is to use the idea of binary numbers. Use the binary represntation to find where is the last occurance of 1 is : thats where we get an odd number when we keep dividing by 2 for the first time. Then make sure that we have atleas that many powers of two for the Jerry strength. This can be attained by taking out corresponding number of terms from the right of binary expresion. For example: binray_num/2^k is same as removing last k terms of binary_num

```python
# cook your dish here
#int(input().strip())
#list(map(int,input().strip().split())
if __name__=="__main__":
    T = int(input().strip())
    for _ in range(T):
        TS=int(input().strip())
        if TS==1:
            print(0)
        else:
            TS_bin = bin(TS)
            last_occurence_of_1 = TS_bin.rfind('1')
            if last_occurence_of_1==2:
                print(0)
            else:
                print(int(TS_bin[:last_occurence_of_1],2))
```
