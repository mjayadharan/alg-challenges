```python
t = int(input().strip())
for a0 in range(t):
    n = int(input())
    num_3_divisors = (n-1)//3
    num_5_divisors = (n-1)//5
    num_15_divisors = (n-1)//15
    sum_3_divisors = 3*((num_3_divisors*(num_3_divisors+1))//2)
    sum_5_divisors = 5*((num_5_divisors*(num_5_divisors+1))//2)
    sum_15_divisors = 15*((num_15_divisors*(num_15_divisors+1))//2)
    print(sum_3_divisors+sum_5_divisors-sum_15_divisors)
```
