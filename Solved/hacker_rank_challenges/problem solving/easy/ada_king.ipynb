```python

if __name__=="__main__":
    T = int(input().strip())
    
    for _ in range(T):
        board_grid = [['X' for j in range(8)] for i in range(8)]
        K = int(input())
        num_pos_occupied = 0
        for i in range(8):
            for j in range(8):
                board_grid[i][j] = '.'
                num_pos_occupied += 1
                if num_pos_occupied >=K:
                    break
            if num_pos_occupied >=K:
                break
        board_grid[0][0] = 'O'
        for row in board_grid:
            print("".join(row))
```
