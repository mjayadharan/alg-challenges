```python

if __name__=="__main__":
    T = int(input().strip())
    
    for _ in range(T):
        N = int(input())
        chef_points = 0 # +ve means chef_wins, -ve means morty win
        morty_points = 0
        for i in range(N):
            A,B = list(map(int,input().strip().split()))
            A = sum(map(int,list(str(A))))
            B = sum(map(int,list(str(B))))
            print(A,B)
            if A>B:
                chef_points += 1
            elif A<B:
                morty_points += 1
            else:
                chef_points +=1
                morty_points +=1
        if chef_points > morty_points:  #chef wins
            print(0,chef_points)  
        elif chef_points < morty_points: #morty wins
            print(1,morty_points)
        else:   #draw
            print(2,chef_points)
```
