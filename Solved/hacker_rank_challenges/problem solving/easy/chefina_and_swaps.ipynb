```python
#### from collections import Counter
# from collections import deque
def find_min_cost(A,B):
    minimium_cost = 0
    A_counter = Counter(A)
    B_counter = Counter(B)
    N = len(A)
    needs_change_indices = []
    for i in range(N):
        if (A_counter.get(A[i]) + B_counter.get(A[i],0))%2 != 0 or\
        (A_counter.get(B[i],0) + B_counter.get(B[i]))%2 != 0:
            return -1
        if A_counter.get(A[i]) > B_counter.get(A[i],0):
            needs_change_indices.append(A[i])
            A_counter[A[i]] -= 1
            try:
                B_counter[A[i]] += 1
            except KeyError:
                B_counter[A[i]] = 1
        if B_counter.get(B[i]) > A_counter.get(B[i],0):
            needs_change_indices.append(B[i])
            B_counter[B[i]] -= 1
            try:
                A_counter[B[i]] += 1
            except KeyError:
                A_counter[B[i]] = 1
    needs_change_indices.sort()
    print(needs_change_indices)
    return sum(needs_change_indices[:len(needs_change_indices)//2])
    

if __name__=="__main__":
    T = int(input().strip())
    
    for _ in range(T):
        N = int(input())
        A = list(map(int,input().strip().split()))
        B = list(map(int,input().strip().split()))
        print(find_min_cost(A,B))
```
