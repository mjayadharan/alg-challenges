```python


# Complete the fairRations function below.
def fairRations(B):
    odd_bread_index=[]
    for i in range(len(B)):
        if B[i]%2!=0:
            odd_bread_index.append(i)
    print(odd_bread_index)
    if len(odd_bread_index)%2!=0:
        return "NO"
    else:
        num_breads_required=0
        for i in range(0,len(odd_bread_index),2):
            num_breads_required+=2*(odd_bread_index[i+1]-odd_bread_index[i])
        return num_breads_required





# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     N = int(input())

#     B = list(map(int, input().rstrip().split()))

#     result = fairRations(B)

#     fptr.write(str(result) + '\n')

#     fptr.close()
```
