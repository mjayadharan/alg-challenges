```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

def jumpingOnClouds(c):
    n = len(c)
    num_steps = 0
    current_cloud=0
    while(current_cloud<n-1):
        if (current_cloud!=n-2) and (c[current_cloud+2]==0):
            num_steps+=1
            current_cloud+=2
        elif (current_cloud==n-2):
            num_steps+=1
            current_cloud+=1
        elif (current_cloud!=n-2) and (c[current_cloud+2]==1):
            num_steps+=1
            current_cloud+=1
    return(num_steps)




# if __name__=="__main__":
#     n = int(input().strip())
#     c = list(map(int,input().strip().split()))
#     print(jumpingOnClouds(c))
```
