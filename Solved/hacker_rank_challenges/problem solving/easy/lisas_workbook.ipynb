```python
# Complete the workbook function below.
def workbook(n, k, arr):
    num_special_problems=0
    chapter_last_page=0
    for i in range(n):
        if arr[i]<chapter_last_page:
            num_chapt_pages = arr[i]//k if arr[i]%k==0 else arr[i]//k +1
            chapter_last_page+= num_chapt_pages
            
#             print("reached skip for i={}, arr[i]={}, chap")
            continue
        else:
            num_chapt_pages = arr[i]//k if arr[i]%k==0 else arr[i]//k +1
            in_pages_without_special = chapter_last_page//k
            problems_passed=k*in_pages_without_special
            for page_num in range((chapter_last_page+1)+in_pages_without_special,chapter_last_page+num_chapt_pages+1):
                if page_num>problems_passed and page_num<=min(problems_passed+k,arr[i]):
                    num_special_problems+=1
                problems_passed+=k
            chapter_last_page+= num_chapt_pages
    return num_special_problems




# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     nk = input().split()

#     n = int(nk[0])

#     k = int(nk[1])

#     arr = list(map(int, input().rstrip().split()))

#     result = workbook(n, k, arr)

#     fptr.write(str(result) + '\n')

#     fptr.close()
```
