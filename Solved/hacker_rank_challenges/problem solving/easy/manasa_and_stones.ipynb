```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the stones function below.
def stones(n, a, b):
    result=[]
    if a==b:
        result.append(a*(n-1))
    else:
        diff= max(a,b)-min(a,b)
        last_stone= min(a,b)*(n-1)
        result.append(last_stone)
        while(last_stone<max(a,b)*(n-1)):
            last_stone+=diff
            result.append(last_stone)
    return result

# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     T = int(input())

#     for T_itr in range(T):
#         n = int(input())

#         a = int(input())

#         b = int(input())

#         result = stones(n, a, b)

#         fptr.write(' '.join(map(str, result)))
#         fptr.write('\n')

#     fptr.close()
```
