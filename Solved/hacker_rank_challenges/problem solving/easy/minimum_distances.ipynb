```python
def minimumDistances(arr):
    dic = {}
    # arr_str = "-".join(arr)
    min_distance=len(arr)
    for el in arr:
        str_el = str(el)
        if str_el in dic:
            dic[str_el]+=1
        else:
            dic[str_el]=1
    if max(dic.values())<2:
        return -1
    else:
        for i in range(len(arr)-1):
            if dic[str(arr[i])]>1:
                for j in range(i+1,len(arr)):
                    if arr[i]==arr[j]:
                        dic[str(arr[i])]-=1
                        new_distance = j-i
                        if new_distance<min_distance:
                            min_distance=new_distance
                        break
    return min_distance
                        



# if __name__=="__main__":
#     n = int(input().strip())
#     arr = [int(A_temp) for A_temp in input().strip().split(' ')]
#     print(minimumDistances(arr))
```
