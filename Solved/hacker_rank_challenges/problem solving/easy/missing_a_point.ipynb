```python
from collections import Counter
if __name__=="__main__":
    T = int(input().strip())
    
    for _ in range(T):
        N = int(input())
        x_cord_list=[0]*(4*N-1)
        y_cord_list=[0]*(4*N-1)
        missing_x = 0
        missing_y = 0
        for i in range(4*N-1):
            x_cord_list[i], y_cord_list[i] = list(map(int,input().split()))
        x_counter = Counter(x_cord_list)
        y_counter = Counter(y_cord_list)
        for x_cord in x_counter:
            if x_counter[x_cord]%2 != 0:
                missing_x = x_cord
                break
        for y_cord in y_counter:
            if y_counter[y_cord]%2 != 0:
                missing_y = y_cord
                break
        print(missing_x, missing_y)
        
```
