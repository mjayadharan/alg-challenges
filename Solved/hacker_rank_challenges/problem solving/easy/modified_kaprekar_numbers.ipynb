```python

# Complete the kaprekarNumbers function below.
def kaprekarNumbers(p, q):
    kapr_exist=False
    for i in range(p,q+1):
        d = len(str(i))
        num_str= str(i*i)
        if (len(num_str)<2):
            if i==1:
                print(i,end=" ")
            continue
        r = int(num_str[len(num_str)//2:])
        l = int(num_str[:len(num_str)//2])
        if r+l==i:
            kapr_exist=True
            print(i,end=" ")
    if not kapr_exist:
        print("INVALID RANGE")

kaprekarNumbers(1,1000)
```
