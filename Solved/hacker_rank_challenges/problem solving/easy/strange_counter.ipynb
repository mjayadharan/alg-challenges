```python
import math

index_finder = lambda s_n: math.floor(math.log(s_n/3 +1)/math.log(2))
findSn= lambda n: 3*(2**n -1)
#
# Complete the strangeCounter function below.
#
def strangeCounter(t):
    if t<3:
        return 4-t
    n = index_finder(t)
    s_n = findSn(n)
    if t==s_n:
        return 1
    else:
        return (3*(2**n)+1)-(t-s_n)


# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     t = int(input())

#     result = strangeCounter(t)

#     fptr.write(str(result) + '\n')

#     fptr.close()
```
