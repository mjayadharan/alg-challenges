```python
import sys

# Complete the minimumBribes function below.
def minimumBribes(n,q):
    num_bribes = 0
    if n==1:
        print(num_bribes)
        return None
    new_q = [el-1 for el in q]
    for i,el in enumerate(new_q):
        if el-i>2:
            print("Too chaotic")
            return None
        for j in range(max(0,el-1),i):
            # print(i,new_q[j],el)
            if new_q[j]>el:
                num_bribes+=1
    print(num_bribes)
    

# if __name__ == '__main__':
#     t = int(input())

#     for t_itr in range(t):
#         n = int(input())

#         q = list(map(int, input().rstrip().split()))

#         minimumBribes(n,q)
```
