```python
def biggerIsGreater(word):
    if len(word)<2:
        return "no answer"
    highest_suffix_index=len(word)-1
#     print("word is :",word)
    for i in range(len(word)-2,-1,-1):
        if word[i]<word[i+1]:
            highest_suffix_index= i+1
#             print("reached here 1",highest_suffix_index,word[highest_suffix_index])
            break
        elif i==0:
            return "no answer"
    #prefix exists
    if highest_suffix_index-1>=0:
        pivot = word[highest_suffix_index-1]
#         print("pivot is : ",pivot)
        swap_element_index= highest_suffix_index
        for i in range(highest_suffix_index,len(word)):
            if word[i]<=pivot:
                swap_element_index = i-1
                break
            elif i==len(word)-1:
                swap_element_index = i
#         if highest_index-2>=0
        new_word = word[:highest_suffix_index-1]+word[swap_element_index]+word[highest_suffix_index:swap_element_index]+pivot
        if swap_element_index<len(word)-1:
            new_word+=word[swap_element_index+1:]
        newest_word = new_word[:highest_suffix_index]+new_word[highest_suffix_index:len(new_word)][::-1]
        return newest_word
                
    #no-prefix exists
    else:
        return "no answer"
```
