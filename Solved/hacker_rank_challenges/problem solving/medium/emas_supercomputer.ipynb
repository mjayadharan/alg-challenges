```python
class EmasComputer():
    def __init__(self,arr):
        self.arr = arr
        self.largest_plus_len = 0 #length of largest plus(this is actually the area of the plus)
        self.second_plus_len = 0 #length of second largest plus(this is actually the are of the plus)
        self.largest_plus_indices = [] #hold lists of max length plus indices 
    
    def plus_with_center(self, ind_i, ind_j, arr=[]):
        if not arr:
            arr = self.arr
        plus_indices = set()
        inPlus = True
        i = 0
        while(inPlus):
            if arr[ind_i][ind_j+i] == "G" and arr[ind_i][ind_j-i] == "G" \
            and arr[ind_i+i][ind_j] == "G" and arr[ind_i-i][ind_j] == "G":
                plus_indices = plus_indices.union({(ind_i+i,ind_j),(ind_i-i,ind_j),\
                                    (ind_i,ind_j+i),(ind_i,ind_j-i)})
                i +=1
            else:
                inPlus = False
        return plus_indices
            
    def find_largest_good_plus(self):
        for ind_i in range(1,len(self.arr)-1):
            for ind_j in range(1,len(self.arr[0])-1):
                plus_indices = self.plus_with_center(ind_i,ind_j)
                if len(plus_indices) > self.largest_plus_len:
                    self.largest_plus_len = len(plus_indices)
                    self.largest_plus_indices = [plus_indices]
                elif len(plus_indices) == self.largest_plus_len:
                    self.largest_plus_indices.append(plus_indices)
    
    def find_second_largest_good_plus(self):
        import copy 
        for set_of_indices in self.largest_plus_indices:
            new_array = copy.deepcopy(self.arr)
            for i,j in set_of_indices:
                new_array[i][j] = "*"
            for el in new_array:
                print("".join(el))
            for ind_i in range(1,len(new_array)-1):
                for ind_j in range(1,len(new_array[0])-1):
                    plus_indices = self.plus_with_center(ind_i,ind_j,new_array)
                    if len(plus_indices) > self.second_plus_len:
                        self.second_plus_len = len(plus_indices)
                    if len(plus_indices) == self.largest_plus_len:
                        return None
    
    def find_plus_product_areas(self):
        self.find_largest_good_plus()
        self.find_second_largest_good_plus()
        return self.largest_plus_len*self.second_plus_len

                
                
        
        

        
```
