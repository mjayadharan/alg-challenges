```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import Counter
# Complete the freqQuery function below.
def freqQuery(queries):
    num_queries = len(queries)
    ans_arr = []
    number_counter=Counter()
    freq_counter = Counter()
    #flag==1: insert element, flag==2: delete one occurent, flag==3: return Ture or False
    for flag,number in queries:
        if flag==1:
            if number_counter.get(number):
                freq_counter[number_counter[number]]-=1
            number_counter.update([number])
            freq_counter.update([number_counter[number]])
        elif flag==2:
            if number_counter.get(number):
                freq_counter[number_counter[number]]-=1
                number_counter[number]-=1
                if number_counter[number]>0:
                    freq_counter.update([number_counter[number]])
        elif flag==3: 
            if freq_counter.get(number):
                ans_arr.append(1)
            else:
                ans_arr.append(0)
    return ans_arr



# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     q = int(input().strip())

#     queries = []

#     for _ in range(q):
#         queries.append(list(map(int, input().rstrip().split())))

#     ans = freqQuery(queries)

#     fptr.write('\n'.join(map(str, ans)))
#     fptr.write('\n')

#     fptr.close()
```
