```python

def longest_com_subseq(seq_1,seq_2):
    """
    Returns the longest common subsequence of two iterable objects
    
    Parameters:
    seq_1(iterable): could be string, list of int
    seq_2(iterable): iterable of the same kind as above 
    
    Return: 
    sub_seq(iterable): longest common substring of seq_1 and seq_2
    
    """
    n_1 = len(seq_1)
    n_2 = len(seq_2)
    
    if n_1 == 0 or n_2 == 0:
        return seq_1[0:0]
    elif seq_1[-1] == seq_2[-1]:
        if type(seq_1) == str:
            return longest_com_subseq(seq_1[:-1],seq_2[:-1])+seq_1[-1]
        elif type(seq_2) == list:
            return longest_com_subseq(seq_1[:-1],seq_2[:-1])+[seq_1[-1]]
    else:
        sub_sequence_1 = longest_com_subseq(seq_1[:-1],seq_2)
        sub_sequence_2 = longest_com_subseq(seq_1,seq_2[:-1])
        if len(sub_sequence_1) > len(sub_sequence_2):
            return sub_sequence_1
        else:
            return sub_sequence_2
        
        
```

```python
string_1 = "WEWOUCUIDGCGTRMEZEPXZFEJWISRSBBSYXAYDFEJJDLEBVHHKS"
string_2 = "FDAGCXGKCTKWNECHMRXZWMLRYUCOCZHJRRJBOAJOQJZZVUYXIC"
# longest_com_subseq("ABCD","ACBAD")
longest_com_subseq(string_1,string_2)
```
