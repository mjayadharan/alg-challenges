```python
# #!/bin/python3

# import math
# import os
# import random
# import re
# import sys
from collections import Counter

def row_sum(i,arr):
    return sum([el for el in arr[i]])

def column_sum(j,arr):
    return sum([arr[i][j] for i in range(len(arr))])

# Complete the organizingContainers function below.
def organizingContainers(container):
    n = len(container)
    container_capacity=[]
    for i in range(n):
        container_capacity.append(row_sum(i,container))
    ball_count_per_type =[]
    for j in range(n):
        ball_count_per_type.append(column_sum(j,container))
    container_counter = Counter(map(str,container_capacity))
    ball_counter = Counter(map(str,ball_count_per_type))
    # print(container_counter,ball_counter,sep="\n")
    if container_counter==ball_counter:
        return "Possible"
    else:
        return "Impossible"
    

# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     q = int(input())

#     for q_itr in range(q):
#         n = int(input())

#         container = []

#         for _ in range(n):
#             container.append(list(map(int, input().rstrip().split())))

#         result = organizingContainers(container)

#         fptr.write(result + '\n')

#     fptr.close()
```
