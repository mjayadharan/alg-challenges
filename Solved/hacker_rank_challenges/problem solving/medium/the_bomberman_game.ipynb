__Useful tip:__ 

1) if n=0,1 nothing happens. 

2) if n even n%2==0: grid is full of bombs

3) grid config same for other n%4 ie, for n%4=3 like 3,7,11 and for n%1 like 5,9,13 etc

```python
#!/bin/python3

import math
import os
import random
import re
import sys

def after_explotion(arr):
    ''' after_explotion(arr)->result_array
    This function takes an arr of config of bombs and returns an array config after explotion.
    For this we assume the resul_array initially had all spots filled with bombs, but only the
    in the arr exploded.
    input-> array of strings
    output-> aray of strings of similar kind as input
    '''
    r = len(arr)
    c = len(arr[0])
    result_array=[['O' for j in range(c)] for i in range(r)] 
    for i in range(r):
        for j in range(c):
            if arr[i][j]=='O':
                result_array[i][j]='.'
                try:
                    result_array[i+1][j]='.'
                except:
                    None
                try:
                    if i>0:
                        result_array[i-1][j]='.'    
                except:
                    None
                try:
                    result_array[i][j+1]='.'
                except:
                    None
                try:
                    if j>0:
                        result_array[i][j-1]='.' 
                except:
                    None
    result_array =["".join(el) for el in result_array]
    return result_array
            



# Complete the bomberMan function below.
def bomberMan(n, grid):
    full_bomb_grid = ['O'*len(grid[0]) for i in range(len(grid))]
    r = len(grid)
    c = len(grid[0])
    # print(full_bomb_grid)
    if n%2==0:
        return full_bomb_grid
    if n==0:
        return grid
    elif n==1:
        return grid
    else:
        grid_3=after_explotion(grid)
        if n==3:
            return grid_3
        grid_5 = after_explotion(grid_3)
        if n==5:
            return grid_5
        # grid_7 = after_explotion(grid_5)
        if n%4==1:
            return grid_5
        elif n%4==3:
            return grid_3





# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     rcn = input().split()

#     r = int(rcn[0])

#     c = int(rcn[1])

#     n = int(rcn[2])

#     grid = []

#     for _ in range(r):
#         grid_item = input()
#         grid.append(grid_item)

#     result = bomberMan(n, grid)

#     fptr.write('\n'.join(result))
#     fptr.write('\n')

#     fptr.close()
```
