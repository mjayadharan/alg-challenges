```python
def hackerrankInString(s):
    required_string="hackerrank"
    if len(s)<len(required_string):
        return "NO"
    else:
        next_index_to_append=0
        for el in s:
            if el==required_string[next_index_to_append]:
                next_index_to_append+=1
            if next_index_to_append==len(required_string):
                return "YES"
        return "NO"       

# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     q = int(input())

#     for q_itr in range(q):
#         s = input()

#         result = hackerrankInString(s)

#         fptr.write(result + '\n')

#     fptr.close()
```
