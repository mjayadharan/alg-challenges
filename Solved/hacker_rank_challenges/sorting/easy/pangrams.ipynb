```python
import math
import os
import random
import re
import sys
import string

# Complete the pangrams function below.
def pangrams(s):
    if len(s)<20:
        return "not pangram"
    lower_alph = list(string.ascii_lowercase)
    # print(lower_alph)
    s = s.lower()
    # print(s)
    for el in s:
        if el in lower_alph:
            lower_alph.remove(el)
            # print(lower_alph)
            if len(lower_alph)==0:
                return "pangram"
    return "not pangram"


# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     s = input()

#     result = pangrams(s)

#     fptr.write(result + '\n')

#     fptr.close()
```
