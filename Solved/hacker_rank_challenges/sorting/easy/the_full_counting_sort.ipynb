```python
def countSort(arr):
    int_indices =[int(el[0]) for el in arr]
    max_element = max(int_indices)
    print(max_element)
    count_array = [0 for i in range(max_element+1)]
    assist_list=[[] for i in range(max_element+1)]
    # print(max_element)
    if max_element==98:
        count_array.append(0)
    for  i in range(len(arr)):
        count_array[int(arr[i][0])]+=1
        if i<len(arr)//2:
            assist_list[int(arr[i][0])].append("-")
        else:
            assist_list[int(arr[i][0])].append(arr[i][1])
    for i in range(len(count_array)):
        for j in range(count_array[i]):
            print(assist_list[i][j],end=' ')
    print()
    # return result_array
```

```python
if __name__ == '__main__':
    file_handler = open('input_text.txt','r')
    n = int(file_handler.readline().strip())

    arr = []

    for _ in range(n):
        arr.append(file_handler.readline().rstrip().split())

    countSort(arr)
```
