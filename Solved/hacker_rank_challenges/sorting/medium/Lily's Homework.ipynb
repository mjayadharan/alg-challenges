**Basic Idea**

- Min distance is obtained when the list is sorted in ascending or descending order.

- So we first sort the list(acending order). 

- Then make a dic with elements in list as key and their index in the list as the value.
- Go through the list and update elements different from the sorted list, increase the num of swaps required and update the dictionary.
- To take care of the descending order of sort, we simply reverse the original list, come up with a dic as in the previous case and calcualte the num of swaps required with respect to the sorted list(ascending). 
- Ouput the min of the two

```python
# Complete the lilysHomework function below.
def lilysHomework(arr):
    n=len(arr)
    if n==1:
        return 0
    num_swaps_forward = 0
    num_swaps_backward = 0
    index_dic = {}
    for i in range(n):
        try:
            index_dic[arr[i]]=i
        except:
            print("something went wrong here")
    arr_backward = arr.copy()
    # print(arr)
    arr_backward.reverse()
    index_dic_backward={}
    for i in range(n):
        try:
            index_dic_backward[arr_backward[i]]=i
        except:
            print("something went wrong here")
    arr_sorted = sorted(arr) 
    for i in range(n):
        min_el = arr_sorted[i]
        # print(min_el)
        if min_el!=arr[i]:
            arr[index_dic[min_el]]=arr[i]
            index_dic[arr[i]]=index_dic[min_el]
            arr[i]=min_el
            num_swaps_forward+=1
            # print(index_dic)
        if min_el!=arr_backward[i]:
            arr_backward[index_dic_backward[min_el]]=arr_backward[i]
            index_dic_backward[arr_backward[i]]=index_dic_backward[min_el]
            arr_backward[i]=min_el
            num_swaps_backward+=1
    return min(num_swaps_forward,num_swaps_backward)
    

# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     n = int(input())

#     arr = list(map(int, input().rstrip().split()))

#     result = lilysHomework(arr)

#     fptr.write(str(result) + '\n')

#     fptr.close()
```
