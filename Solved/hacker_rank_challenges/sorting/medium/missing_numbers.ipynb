```python
def missingNumbers(arr, brr):
    brr_dic ={}
    for el in brr:
        try:
            brr_dic[el]+=1
        except:
            brr_dic[el]=1
    arr_dic={i:0 for i in brr_dic.keys()}
    for el in arr:
        arr_dic[el]+=1
#     print(arr_dic)
    missing_numbers=[]
    for number in brr_dic:
        if brr_dic[number]>arr_dic[number]:
            missing_numbers+=[number]*(brr_dic[number]-arr_dic[number])
            # print(missing_numbers)
    missing_numbers.sort()
    return missing_numbers
```
