```python
# Complete the balancedSums function below.
def balancedSums(arr):
    if len(arr)==1:
        return "YES"
    # arr.sort()
    n = len(arr)
    # left_pivot = n//2
    # right_pivot = n//2
    left_sum=0
    right_sum=sum(arr)
    for i in range(0,n):
        left_sum+=0 if i==0 else arr[i-1]
        # right_sum+=0 if i==len(arr)-1 else sum(arr[i+1:])
        right_sum-= arr[i]
        if left_sum==right_sum:
            return "YES"
    return "NO"
        

# if __name__ == '__main__':
#     fptr = open(os.environ['OUTPUT_PATH'], 'w')

#     T = int(input().strip())

#     for T_itr in range(T):
#         n = int(input().strip())

#         arr = list(map(int, input().rstrip().split()))

#         result = balancedSums(arr)

#         fptr.write(result + '\n')

#     fptr.close()
```
