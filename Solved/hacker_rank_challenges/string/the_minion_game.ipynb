```python
def minion_game(s):
    # your code goes here
    vowels = "AEIOU"
    stuart_score = 0
    kevin_score = 0
    for i,letter in enumerate(s):
        if letter in vowels:
            kevin_score+=len(s)-i
        else:
            stuart_score+=len(s)-i
    if stuart_score==kevin_score:
        print( "Draw")
    else:
        print("Kevin {}".format(kevin_score)) if kevin_score>stuart_score else print("Stuart {}".format(stuart_score))

# if __name__ == '__main__':
#     s = input()
#     minion_game(s)
```
